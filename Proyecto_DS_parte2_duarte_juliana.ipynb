**Introducción y descripción de datos**

Este conjunto de datos contiene registros detallados de los eventos olímpicos y de los atletas que participaron en ellos desde el año 1896 hasta 2016. Cada fila en el conjunto de datos representa a un atleta individual e incluye varios atributos relacionados con su información personal, el país que representan, los detalles del evento olímpico en el que participaron y los resultados que lograron. A continuación, se describe cada columna incluida en el conjunto de datos:

- Nombre: El nombre completo del atleta.
- Género: El género del atleta, generalmente representado como 'M' para masculino y 'F' para femenino.
- Edad: La edad del atleta en el momento de su participación en los
Juegos Olímpicos.
- Altura: La altura del atleta en centímetros.
- Peso: El peso del atleta en kilogramos.
- Equipo: El nombre del equipo o delegación que el atleta representa, usualmente corresponde a su país.
- Código del País: El código de tres letras que representa el país asociado al atleta (por ejemplo, USA para Estados Unidos, GBR para Gran Bretaña).
- Juego: El evento olímpico o deporte específico en el que compitió el atleta.
- Año: El año en que el atleta participó en los Juegos Olímpicos.
- Temporada: La temporada de los Juegos Olímpicos (ya sea Verano o Invierno).
- Ciudad: La ciudad sede donde se llevaron a cabo los Juegos Olímpicos.
- Deporte: La categoría deportiva general en la que compitió el atleta (por ejemplo, Atletismo, Natación, Gimnasia).
- Evento: El evento específico dentro del deporte en el que compitió el atleta (por ejemplo, 100m Libre, Salto de Longitud, Maratón).
- Medalla: El tipo de medalla que ganó el atleta, si ganó alguna (Oro, Plata, Bronce). Si el atleta no ganó una medalla, este campo estará vacío.

```python
#reemplazar nulos automaticamente segun conveniencia (promedio o mediana según el desvio estandar) (todas las numericas, menos el año, no le veo mucho sentido)
columnas= [ 'Age', 'Height', 'Weight']

for col in columnas:
  media = df_olimpiadas[col].mean()
  mediana = df_olimpiadas[col].median()
  std = df_olimpiadas[col].std() # del desvio estandar depende si se reemplaza por la media o por la mediana. A veces los extremos afectan mucho a la media por lo que se toma la mediana en esos casos.
  if abs(media-mediana) <= std:
    df_olimpiadas[col].fillna(media, inplace=True)
  else:
    df_olimpiadas[col].fillna(mediana, inplace=True)

df_olimpiadas_sin_outl = pd.DataFrame()  # DataFrame vacío para almacenar los nuevos datos
```

**Matplotlib - 1**

```python
#Agrupo el rango de edad segun interes para analizarlo por segmento y que sea mas significativa la cantidad en cada uno de ellos en comparación con el otro.
def clasificar_edad(Age):
    if Age < 18:
        return 'Menor de 18'
    elif 18 <= Age <= 25:
        return '18-25'
    elif 26 <= Age <= 35:
        return '26-35'
    elif 36 <= Age <= 71:
        return '36-71'
    else:
        return 'Mayor de 71 u otros'

df_olimpiadas['Rango_Edad'] = df_olimpiadas['Age'].apply(clasificar_edad) #aplico la función a la columna edad, y se guarda en una nueva columna que se llama rango edad, para no pisar la información existente.

conteo_ages= df_olimpiadas['Rango_Edad'].value_counts() #cuento cuantos participantes estan en cada rango

#realizo el grafico
plt.figure(figsize=(10, 10))
plt.pie(conteo_ages, labels=conteo_ages.index, autopct='%1.1f%%', colors= sns.color_palette('dark'),startangle=90)
plt.title('Distribución de edades')
plt.show()
```

**Conclusión y analisis del grafico:**

Una vez visualizado el grafico, se observa que efectivamente el rango de edad mas comun es el de "18 a 25 años", si bien se esperaba su resultado, es necesario el analisis para conocer la muestra con la que se esta trabajado y analizando. Por otro lado los menores de edad no constituyen el grupo con menor porcentaje, ya que el rango etario de 36 a 71 años es aun menor con un 3,9%.
Finalmente se destaca la edad maxima de participación dado que realizando un "describe", se observa como valor maximo en la columna de edad 71 años. En base a ello se segmento el ultimo rango de edad desde los 36 a los 71 años. Si bien estan presentes son la minoria.


**Seaborn - 1**

```python
#realizo el grafico en el cual veo la existencia de outliers en ambos boxplots. Por lo tanto para comprander mejor el comportamiento general se realiza a continuación sin los mismos y en función de ello se obtendran conclusiones.
sns.set(style='whitegrid')
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_olimpiadas, x='Sex', y='Height', palette={'M': 'blue', 'F': '#C71585'})
plt.title('Distribución de Altura por Género sin Outliers')
plt.show()
```

```python
for sex in df_olimpiadas['Sex'].unique():  # Necesito que itere sobre ambos generos porque cada uno tiene su intervalo de confianza y medidas significativas, si me toma el general para ambos no va a tener sentido el grafico.
    # Calcular cuartiles y límites
    Q1 = df_olimpiadas[df_olimpiadas['Sex'] == sex]['Height'].quantile(0.25)
    Q3 = df_olimpiadas[df_olimpiadas['Sex'] == sex]['Height'].quantile(0.75)
    IQR = Q3 - Q1 #la resta me da el percentil 50
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Datos pero sin los outliers
    df_sin_outl_sex = df_olimpiadas[(df_olimpiadas['Height'] >= limite_inferior) &
                                     (df_olimpiadas['Height'] <= limite_superior) &
                                     (df_olimpiadas['Sex'] == sex)]

    df_olimpiadas_sin_outl = pd.concat([df_olimpiadas_sin_outl, df_sin_outl_sex])  # Concatenar resultados


#realizo el grafico ahora sin los outliers
sns.set(style='whitegrid')
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_olimpiadas_sin_outl, x='Sex', y='Height', palette={'M': 'blue', 'F': '#C71585'})
plt.title('Distribución de Altura por Género sin Outliers')
plt.show()
```

**Conclusión y analisis del grafico:**

Evaluando la media (Cuartil 50) de ambos sexos se ve que la de los hombres es de 175,3 mientras que la de las mujeres es de 169. Por lo contrario, poseen mayor dispersion las mujeres (se estima que esta relacionado con factores del crecimiento y los diferentes deporte ya que una mujer que juega al basquet tiene significativamente mas altura que una que realiza gimnasia artistica, no asi en los hombres por lo que demuestra el grafico, estan mas compactos), ademas se puede observar en la desviación estandar, ya que me indica que tan lejos estan los percentiles de mi promedio o percentil 50.
Luego, se observa que el boxplot de las mujeres esta mas equilibrado en el centro (con la mediana cerca del rango medio entre Q1 y Q3) en comparacion con el boxplot de los hombres, el cual esta mas proximo al percentil 25 o primer cuartil. Finalmente se estimaba que el percentil 75 de las mujeres es igual al percentil 25 de los hombres, sin embargo por muy poco no es asi, ya que el percentil 75 de las mujeres es de 175,3 mientras que el 25 de los hombres es de 175 (Verificado con exactitud a continuación)

**Conclusión y analisis del método Regresión opción 1**

Entreno con un modelo Stats, el cual me brinda mayor cantidad de datos para entender el problema. Realizando la regresión sobre las variables medallas y edad, podemos concluir en que:
- R2 (R cuadrado): Me informa que tan bien se ajustan los datos a mi linea. En este caso solo toma el 0,2% de mis datos, es muy malo, no esta entendiendo mi linea. Generalmente cuando le agrego mas variabilidad a mi modelo mejoro este valor y ajusto el resultado. Buscamos que este sea lo mas alto posible.
- P-val: Indica que tan significativa es esa variable para ese modelo. En este caso es menor que 0.05 por lo que indica que es muy significativa y es logico ya que solo le otorgue una variable.
- F-statistic: Nos indica cual es la significancia global del modelo. Deberia ser un numero bajo, como no sucede esto ya que es un numero alto (106.5) es otro signo de alarma.
- Kurtosis: La misma no esta cerca de 3 ya que es dificil normalizar la cantidad de metallas ganadas debido a la nueva incorporación de paises, deportes, competidores, etc en casa instancia.
- Skew: Indica la asimetria, que tan simetrico es de cada lado la distribución, en este caso no esta totalmente cerca de 0 por lo que no es simetrica del todo.
- Intervalo de confianza de la pendiente: Nos idica cual es mi minimo y mi maximo. Se encuentra en conflicto cuando uno de ellos es negativo y el otro es positivo, ya que indica que en algun momento va a pasar por cero mi X y eso convertiria mi recta en una constante. Por lo que en caso de ocurrir eso se deben descartar esas variables. En este caso ambos estan con valores negativos por lo que estaria correcto.

Por lo tanto y luego de analizar cada una de las variables anteriormente, descarto este metodo como viable para el analisis. A continuación se le incorporan mas variables independientes para poder mejorar el entendimiento de esta recta en función de los datos otorgados.

**Conclusión método KNN (Metodo 5 y 6)**

El valor de K me indica cuantos "vecinos" debo mirar y tener en cuenta para analizar la cercania y clasificar. Generalmente si yo le doy un k=1 el modelo tiende a sobreaprender (overfiting) porque va caso por caso viendo vecino por vecino, por lo contrario cuando el k es muy grande no aprender (underfiting) simplemente traza una linea donde considera mejor para dividir los datos. Debido a estos factores es necesario ver el K ideal para cada caso

La técnica de vecinos cercanos es útil en problemas de clasificación donde las relaciones entre los datos no son lineales, ya que se adapta de manera flexible al contexto y las particularidades de los elementos a evaluar. Este enfoque es especialmente ventajoso en el ámbito de los deportes de alto rendimiento, donde las características de los atletas suelen desviarse de los parámetros comunes.

En este contexto, los métodos de regresión suelen ser menos eficientes. La regresión busca modelar relaciones lineales o de tendencias predecibles entre variables, lo cual funciona bien cuando los datos presentan una estructura regular. Sin embargo, en el deporte de alto rendimiento, es común encontrar variaciones que no siguen una tendencia lineal. Por ejemplo, los atletas pueden tener contexturas físicas muy por encima de la media, edades que están en los extremos del espectro deportivo, y variables adicionales como nacionalidad, ubicación geográfica, y género, que afectan el rendimiento de formas no lineales.

**Conclusión final**

La técnica de vecinos cercanos es la herramienta mas eficaz para el análisis de desempeño en deportes de alto rendimiento, ya que permite clasificar atletas según factores complejos y no lineales. En este ámbito, características como preparación, experiencia, manejo de presión, y ciertos aspectos físicos varían considerablemente entre individuos, por lo que es fundamental contar con un modelo que se adapte a estas particularidades y logre comprenderlas. A diferencia de los métodos de regresión, que buscan relaciones lineales y son menos adecuados para reflejar la realidad diversa y cambiante de los deportes, el enfoque de vecinos cercanos clasifica en función de similitudes multidimensionales, capturando lo mejor de estas variaciones.
